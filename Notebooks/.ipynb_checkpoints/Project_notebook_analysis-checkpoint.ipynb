# Loading the cleaned dataframes with the cleaning function

```python
import pandas as pd

def clean_data(df_demo, df_experiment_clients, df_web_data_pt_1, df_web_data_pt_2):
    # Replace 'U' values with 'Unknown' and drop rows with 'X' values
    df_demo_cleaned = df_demo.replace({'gendr': {'U': 'Unknown'}})
    df_demo_cleaned = df_demo_cleaned[df_demo_cleaned['gendr'] != 'X']

    # Drop rows with both 'U' and 'X' values
    df_demo_dropped = df_demo[(df_demo['gendr'] != 'U') & (df_demo['gendr'] != 'X')]
    
    # Clean df_experiment_clients
    df_experiment_clients_cleaned = df_experiment_clients.copy()
    df_experiment_clients_cleaned['Variation'] = df_experiment_clients_cleaned['Variation'].fillna('unknown_experiment')
    
    # Combine df_web_data_pt_1 and df_web_data_pt_2
    df_web_data_combined = pd.concat([df_web_data_pt_1, df_web_data_pt_2], ignore_index=True)
    df_web_data_combined.drop_duplicates(inplace=True)
    
    # Merge df_demo_cleaned and df_experiment_clients_cleaned
    df_merged_demo_exp = pd.merge(df_demo_cleaned, df_experiment_clients_cleaned, on='client_id', how='inner')
    
    # Merge df_merged_demo_exp with df_web_data_combined and change datetime type
    df_final_merged = pd.merge(df_merged_demo_exp, df_web_data_combined, on='client_id', how='left')
    df_final_merged['date_time'] = pd.to_datetime(df_final_merged['date_time'])
    
    return df_demo_dropped, df_demo_cleaned, df_experiment_clients_cleaned, df_web_data_combined, df_merged_demo_exp, df_final_merged
```

# Hypothesis testing of the  completion rate (2-sample proportion z-test)

Test used: Two-sample proportion z-test. This test is appropriate for comparing the proportions of two independent groups, in this case, the completion rates of the Test and Control groups.

Null Hypothesis (H0): There is no difference in completion rates between the Test and Control groups.

Alternative Hypothesis (H1): There is a difference in completion rates between the Test and Control groups.

significance level - 0.05

If the p-value obtained from the z-test is less than 0.05, we'll reject the null hypothesis and conclude that there is a statistically significant difference in completion rates between the Test and Control groups.

```python
from statsmodels.stats.proportion import proportions_ztest

# Count of confirmations and starts for Control and Test groups
successes = np.array([confirm_control, confirm_test])
trials = np.array([start_control, start_test])

# Performing two-sample proportion z-test
z_stat, p_value = proportions_ztest(successes, trials, alternative='two-sided')

# z-statistic and p-value
print("Z-statistic:", round(z_stat, 2))
print("P-value:", round(p_value, 2))

# Comparing p-value with significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a statistically significant difference in completion rates. ")
else:
    print("Accept the null hypothesis: There is no statistically significant difference in completion rates.")
```

# KPI 2: Time Spent on Each Step

# 1. Mean total process time in the test and control groups 

```python
# Filtering DataFrame for test and control groups
df_test = df_final_merged[df_final_merged['Variation'] == 'Test']
df_control = df_final_merged[df_final_merged['Variation'] == 'Control']

# Define a function to calculate average time spent on each step
def calculate_average_time(df):
    # Sort the DataFrame by client_id, visit_id, date_time
    df_sorted = df.sort_values(by=['client_id', 'visit_id', 'date_time'])

    # Remove duplicate rows
    df_unique = df_sorted.drop_duplicates(subset=['client_id', 'visit_id', 'process_step', 'date_time'])

    # Exclude error steps
    df_valid_steps = df_unique[~df_unique['process_step'].str.startswith('error')]

    # Keep only the last "start" time for each client
    df_last_start = df_valid_steps[df_valid_steps['process_step'] == 'start'].groupby('client_id').tail(1)

    # Keep only the last "confirm" time for each client
    df_last_confirm = df_valid_steps[df_valid_steps['process_step'] == 'confirm'].groupby('client_id').tail(1)

    # Merge last start and last confirm data
    df_last_start_confirm = pd.concat([df_last_start, df_last_confirm])

    # Calculate time differences between consecutive steps for each visit
    df_last_start_confirm['time_diff'] = df_last_start_confirm.groupby(['client_id', 'visit_id'])['date_time'].diff()

    # Calculate average time spent on each step
    average_time_spent = df_last_start_confirm.groupby('process_step')['time_diff'].mean()

    return average_time_spent

# average time spent for test and control groups
average_time_spent_test = calculate_average_time(df_test)
average_time_spent_control = calculate_average_time(df_control)

print("Average Time Spent for Test Group:")
print(average_time_spent_test)

print("Average Time Spent for Control Group:")
print(average_time_spent_control)
```

# KPI 3: Error rate comparison for test and control samples

```python
# Calculating the count of backward steps for control and test groups
backward_control = df_final_merged[(df_final_merged['Variation'] == 'Control') & (df_final_merged['process_step'] != 'start')].groupby(['client_id', 'visitor_id', 'visit_id']).size().reset_index(name='count').shape[0]
backward_test = df_final_merged[(df_final_merged['Variation'] == 'Test') & (df_final_merged['process_step'] != 'start')].groupby(['client_id', 'visitor_id', 'visit_id']).size().reset_index(name='count').shape[0]

# Calculating the total number of steps for control and test groups
total_steps_control = df_final_merged[df_final_merged['Variation'] == 'Control'].groupby(['client_id', 'visitor_id', 'visit_id']).size().shape[0]
total_steps_test = df_final_merged[df_final_merged['Variation'] == 'Test'].groupby(['client_id', 'visitor_id', 'visit_id']).size().shape[0]

# Calculate error rates for control and test groups
error_rate_control = round(backward_control / total_steps_control, 2)
error_rate_test = round(backward_test / total_steps_test, 2)

# error rates
print("Error rate for control group:", error_rate_control)
print("Error rate for test group:", error_rate_test)

# Check if error rates are the same for control and test groups
if error_rate_control == error_rate_test:
    print("Error rates are the same for control and test groups.")
else:
    print("Error rates are not the same for control and test groups.")
```
